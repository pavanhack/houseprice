```python
x=df.drop(columns=['HOUSEPRICE','MAINTENANCE (MONTHLY)','TOTAL FLOORS'])
y=df['HOUSEPRICE']
```

```python
df.dtypes
```

```python
numvar=df.select_dtypes(include=['int64','float64']).columns
numvar
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest =  train_test_split(x,y,test_size=0.3,random_state=102)
```

```python
catvar=df.select_dtypes(include=['object']).columns
catvar
```

```python

catego=['TYPE', 'FURNISHING', 'CONSTRUCTION STATUS', 'LISTED BY', 'FACING',
       'LOCALITY', 'STATE']
from catboost import CatBoostRegressor

catboo = CatBoostRegressor(iterations=750,
                            random_state = 2021, od_type = 'Iter',
                            eval_metric="R2",learning_rate=0.009,depth=6,l2_leaf_reg=5,bagging_temperature=2
                            ,border_count=255,grow_policy='Lossguide',max_leaves=650)
catboo.fit(xtrain, np.log(ytrain),cat_features=catego,eval_set=(xtest, np.log(ytest)),plot=True)
```

```python

from sklearn.metrics import r2_score
x_pred = catboo.predict(xtrain)
print(r2_score(np.log(ytrain),x_pred))


y_pred = catboo.predict(xtest)
r2_score(np.log(ytest),y_pred)
```

```python
import pickle 
filename = 'olxprediction.pkl'
pickle.dump(catboo,open(filename,'wb'))
```
